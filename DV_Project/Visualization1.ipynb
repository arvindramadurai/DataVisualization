```python
Game2=pd.read_csv('mergednew1.csv',index_col = False )
Game2.fillna(0)
mylist = list(set(Game2['Genre']))
mylist
```

```python
def get_data(plot_name):
    racing_arr = []
    # res = DataFrame(columns=('Year', 'Global_Sales')
    for index,row in Game2.iterrows():
        if ((row['Year']) > 0 and row['Genre'] == plot_name):
            racing_arr.append([np.int(row['Year']),row['Global_Sales']])
    df = pd.DataFrame(racing_arr)
    abc = df.groupby([0], sort=True).sum()
    return abc
a = get_data('Racing')
```

```python
import plotly.plotly as py
def plot_graph(plot_name):
    abc = get_data(plot_name)
    data=[go.Scatter(x=abc.index.values,
             y=abc[1],
             name='Peak Sales',
                     line = dict(
                    color =('rgb(12,205,20)'),
                     width = 4)),
         ]
    layout= go.Layout(title= 'Global Sales(1982-2016) as per Genre',
            xaxis= dict(
            title= 'Year',
            ),
            yaxis=dict(
            title= 'Global Sales',
            ),
            showlegend= True
    )
    fig = go.Figure(data=data, layout=layout)
    p = py.iplot(fig)
    return p
plot_graph('Racing')
```

```python
import ipywidgets as widgets
s = widgets.Dropdown(
    options=mylist,
    value='Racing',
    description='Genre:',
    disabled=False,
)
widgets.interact(plot_graph,plot_name=s)
```

```python
def get_data(plot_name):
    Xbox_arr = []
    # res = DataFrame(columns=('Year', 'Global_Sales')
    for index,row in Game2.iterrows():
        if ((row['score']) > 0 and row['platform'] == plot_name):
            Xbox_arr.append([np.int(row['score']),row['NA_Sales']])
    xbox = pd.DataFrame(Xbox_arr)
    xboxvalue = xbox.groupby([0], sort=True).sum()
    return xboxvalue
console = get_data('Xbox')
```

```python
import plotly.plotly as py
def plot_graph(plot_name):
    xboxvalue = get_data(plot_name)
    data1=[go.Bar(x=xboxvalue.index.values,
             y=xboxvalue[1],
             name='Ratings')]
    layout= go.Layout(title= 'Game Sales based on Ratings in North America',
            xaxis= dict(
            title= 'Ratings',
            ),
            yaxis=dict(
            title= 'North America Sales',
            ),
            showlegend= True
    )
    fig = go.Figure(data=data1, layout=layout)
    px = py.iplot(fig)
    return px

import ipywidgets as widgets
s = widgets.ToggleButtons(
    options=mylist2,
    value='Xbox',
    description='Platform:',
    disabled=False,
)
widgets.interact(plot_graph,plot_name=s)
#plot_graph('Xbox')
```
