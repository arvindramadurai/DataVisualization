```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
Game2=pd.read_csv('mergednew1.csv')
```

```python
Game2.head()
```

```python
scorephrase=Game2.loc[Game2['score_phrase']=='Amazing']
scorephrase_count=scorephrase['Genre'].value_counts()
abc=sns.barplot(scorephrase_count.index,scorephrase_count.values)
abc.set(xlabel='Genre', ylabel='Count', title='Total count of "Amazing" Games in each Genre')
# plt.show()
plt.grid(False)
plt.xticks(rotation=90)
plt.show()
```

```python
w_platform=Game2['Platform'].value_counts()
top_10=w_platform[:10]
w=sns.barplot(top_10.index,top_10.values)
w.set(xlabel='Platforms', ylabel='Game Count', title='Top 10 Platforms')
plt.xticks(rotation=90)
plt.grid(False)
plt.show()
```

```python
scorephrase=Game2['score_phrase']
scorephrase=scorephrase.value_counts()
#sns.reset_orig()
sns.set_style('white')
x=sns.pointplot(scorephrase.index,scorephrase.values)
x.set(xlabel='Score Phrase', ylabel='Game Count', title='Game counts based on different Score phrases')
plt.grid(False)
#sns.axes_style('white')
sns.set(rc={'axes.facecolor':'white','figure.facecolor':'white'})
plt.xticks(rotation=90)
plt.show()
```
